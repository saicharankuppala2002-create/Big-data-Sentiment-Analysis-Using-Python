```python
# ==========================================
# BIG DATA SENTIMENT ANALYSIS PROJECT
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from textblob import TextBlob
from wordcloud import WordCloud
import re


# Load CSV File
df = pd.read_csv('/content/test.csv', encoding='latin1')

# Display First 5 Rows
print("FIRST 5 RECORDS")
print(df.head())


# Remove Null Values
df.dropna(inplace=True)

# Remove Duplicate Rows
df.drop_duplicates(inplace=True)

# Convert Text to Lowercase
df['text'] = df['text'].str.lower()

# Remove Special Characters
def clean_text(text):
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'[^A-Za-z\s]', '', text)
    return text

df['Cleaned_Review'] = df['text'].apply(clean_text)

print("\nDATA CLEANING COMPLETED")

# Sentiment Function
def analyze_sentiment(text):

    analysis = TextBlob(str(text))

    polarity = analysis.sentiment.polarity

    if polarity > 0:
        return 'Positive'

    elif polarity < 0:
        return 'Negative'

    else:
        return 'Neutral'

# Apply Sentiment Analysis
df['Sentiment'] = df['Cleaned_Review'].apply(analyze_sentiment)

print("\nSENTIMENT ANALYSIS COMPLETED")


sentiment_counts = df['Sentiment'].value_counts()

print("\nSENTIMENT COUNTS")
print(sentiment_counts)


plt.figure(figsize=(8,5))

sentiment_counts.plot(
    kind='bar',
    fontsize=12
)

plt.title('Sentiment Analysis Result', fontsize=16)
plt.xlabel('Sentiment Type', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)

plt.xticks(rotation=0)

plt.savefig('bar_graph.png')

plt.show()


plt.figure(figsize=(7,7))

sentiment_counts.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel('')

plt.title('Sentiment Distribution')

plt.savefig('pie_chart.png')

plt.show()


all_words = ' '.join(df['Cleaned_Review'])

wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color='white'
).generate(all_words)

plt.figure(figsize=(12,6))

plt.imshow(wordcloud)

plt.axis('off')

plt.title('Most Frequent Words')

plt.savefig('wordcloud.png')

plt.show()


df.to_csv('final_sentiment_output.csv', index=False)

print("\nFINAL OUTPUT FILE SAVED")


positive_reviews = len(df[df['Sentiment'] == 'Positive'])
negative_reviews = len(df[df['Sentiment'] == 'Negative'])
neutral_reviews = len(df[df['Sentiment'] == 'Neutral'])

total_reviews = len(df)

print("\nPROJECT SUMMARY")
print("========================")
print("Total Reviews :", total_reviews)
print("Positive Reviews :", positive_reviews)
print("Negative Reviews :", negative_reviews)
print("Neutral Reviews :", neutral_reviews)


print("\nTOP 10 CLEANED REVIEWS")
print(df['Cleaned_Review'].head(10))


report = f"""

BIG DATA SENTIMENT ANALYSIS REPORT
==================================

Total Reviews : {total_reviews}

Positive Reviews : {positive_reviews}

Negative Reviews : {negative_reviews}

Neutral Reviews : {neutral_reviews}

Project Completed Successfully

"""

with open("project_report.txt", "w") as file:
    file.write(report)

print("\nPROJECT REPORT GENERATED")


print("\nBIG DATA SENTIMENT ANALYSIS PROJECT COMPLETED SUCCESSFULLY")
```
